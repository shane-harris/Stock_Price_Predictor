```
df_sent = pd.get_dummies(df_copy, columns=['sentiment'])
df_sent.drop(['Unnamed: 0'], axis=1)
```

```
# df_agg = {'Ticker': 'first','text': df_sent.agg(['text']), 'title': df_sent.agg(['title']),  'sentiment_Negative': 'sum', 'sentiment_Neutral': 'sum', 'sentiment_Positive': 'sum'}
# df_agg = {'Ticker': 'first', 'sentiment_Negative': 'sum', 'sentiment_Neutral': 'sum', 'sentiment_Positive': 'sum'}
df_agg = {'date':'first','Ticker': 'first','text': 'first', 'title': 'first', 'sentiment_Negative': 'sum', 'sentiment_Neutral': 'sum', 'sentiment_Positive': 'sum'}
# df_txt_titl = pd.DataFrame({'df_sent':title, 'df_sent':text})
# text = pd.Series(df_sent['text'])
# df_new = df_sent.groupby(['Ticker','date'])['text','title'].apply(''.join).aggregate(df_agg).reset_index()
# df_new = df_sent.groupby(['Ticker','date'])[df_sent['text'], df_sent['title']].apply(' '.join).reset_index()
# df_new = df_sent.groupby(['Ticker','date']).apply(lambda x: x.combine_first(df_sent['text'], axis=1).combine_first(df_sent['title'], axis=1))

df_new = df_sent.groupby(['Ticker','date']).aggregate(df_agg)
```

```
df_new.reset_index(inplace=True,drop=True)

df_new
```

```
df_new.iloc[0, df_new.columns.get_loc('sentiment_Neutral')]
```

```

df_new['sentiment_Negative'] = df_new['sentiment_Negative'].astype(float)
df_new['sentiment_Neutral'] = df_new['sentiment_Neutral'].astype(float)
df_new['sentiment_Positive'] = df_new['sentiment_Positive'].astype(float)
# total =0
# for index, dfs in df_new.iterrows():
#   total = dfs['sentiment_Negative'] + dfs['sentiment_Neutral'] + dfs['sentiment_Positive']
#   # print(total)
#   # print(dfs['sentiment_Negative']/total)
#   neg = dfs['sentiment_Negative']/total
#   # print(neg)
#   neu = dfs['sentiment_Neutral']/total
#   pos = dfs['sentiment_Positive']/total
#   # df_new.at[dfs, 'sentiment_Negative'] = neg
#   df_new.set_value(dfs,'sentiment_Neutral',neu)
#   # df_new.set_value(dfs,'sentiment_Positive',pos)
#   # print(dfs['sentiment_Negative'])

for i in range(len(df_new)):
  df_new.loc[i,'sentiment_Negative'] = (df_new.loc[i,'sentiment_Negative'])/(df_new.loc[i,'sentiment_Negative']+df_new.loc[i,'sentiment_Neutral']+df_new.loc[i,'sentiment_Positive'])
  df_new.loc[i,'sentiment_Neutral'] = (df_new.loc[i,'sentiment_Neutral'])/(df_new.loc[i,'sentiment_Negative']+df_new.loc[i,'sentiment_Neutral']+df_new.loc[i,'sentiment_Positive'])
  df_new.loc[i,'sentiment_Positive'] = (df_new.loc[i,'sentiment_Positive'])/(df_new.loc[i,'sentiment_Negative']+df_new.loc[i,'sentiment_Neutral']+df_new.loc[i,'sentiment_Positive'])

# df_new.replace(to_replace=df_new['sentiment_Negative'], value=(df_new['sentiment_Negative']/(df_new['sentiment_Negative']+df_new['sentiment_Neutral']+df_new['sentiment_Positive'])))
# df_new.replace(to_replace=df_new['sentiment_Neutral'], value=(df_new['sentiment_Neutral']/(df_new['sentiment_Negative']+df_new['sentiment_Neutral']+df_new['sentiment_Positive'])))
# df_new.replace(to_replace=df_new['sentiment_Positive'], value=(df_new['sentiment_Positive']/(df_new['sentiment_Negative']+df_new['sentiment_Neutral']+df_new['sentiment_Positive'])))


df_new.head(500)
```

```
text_list = list(df_new['text'])
title_list = list(df_new['title'])
```

```
vectorizer1 = TfidfVectorizer(ngram_range=(1,1), max_features=500)
vectorizer2 = TfidfVectorizer(ngram_range=(1,1), max_features=500)
```

```
text_x = vectorizer1.fit_transform(text_list)
title_x = vectorizer2.fit_transform(title_list)
```

```
print(vectorizer2.get_feature_names())
```

```
df1 = pd.DataFrame(text_x.toarray(), columns=vectorizer1.get_feature_names())
df2 = pd.DataFrame(title_x.toarray(), columns=vectorizer2.get_feature_names())
```

```
sub_df = df_new[['Ticker','date','sentiment_Negative','sentiment_Neutral','sentiment_Positive']]
```

```
sub_df
```

```
new_df = pd.concat([sub_df, df1],axis=1)
```

```
new_df1 = pd.concat([new_df,df2], axis=1)
```
